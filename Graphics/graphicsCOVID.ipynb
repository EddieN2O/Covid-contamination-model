```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import os 

os.getcwd()
```

# Datasets

```python
bdbrasilcases = pd.read_csv('https://raw.githubusercontent.com/elhenrico/covid19-Brazil-timeseries/master/confirmed-new.csv',sep = ',')
bdbrasildeaths = pd.read_csv('https://raw.githubusercontent.com/elhenrico/covid19-Brazil-timeseries/master/deaths-new.csv',sep = ',')
```

```python
d041620 = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/04-16-2020.csv',sep=',')
d041720 = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/04-17-2020.csv',sep=',')
```

# Brasil

```python
x = range(1, bdbrasilcases.shape[1]-1)
```

## Casos

```python
bdBRcases = bdbrasilcases.iloc[:,2:bdbrasilcases.shape[1]]
y = bdBRcases.apply(np.sum,axis = 0)
plt.plot(x,y,marker='.')
plt.title('NÃºmero de casos confirmados por COVID-19 no Brasil. Dia 0 |---| {}'.format(max(x)))
plt.xlabel('Dia do contagio')
plt.ylabel('Numero de Casos')
plt.grid(True)

plt.savefig('numbercasesBR.png')
```

```python
bdBRcases = bdbrasilcases.iloc[:,2:bdbrasilcases.shape[1]]

cumulative_cases_br = np.zeros(max(x))
k = 0
for i in range(0,(max(x))):
    cumulative_cases_br[i] = bdBRcases.iloc[0,i] + k
    k = cumulative_cases_br[i]

plt.plot(cumulative_cases_br,marker='.')
plt.title('NÃºmero de casos confirmados no Brasil (Acumulado). Dia 0 |---| {}'.format(max(x) - 1))
plt.xlabel('Dia do Contagio')
plt.ylabel('Numero de casos')
plt.grid(True)

plt.savefig('numbercasesBRcum.png')
```

## Mortes

```python
bdBRdeaths = bdbrasildeaths.iloc[:,2:bdbrasildeaths.shape[1]]

y = bdBRdeaths.apply(np.sum,axis = 0)
plt.plot(x,y,marker='.')
plt.title('NÃºmero de mortes confirmadas por COVID-19 no Brasil. Dia 0 |---| {}'.format(max(x)))
plt.xlabel('Dia do contagio')
plt.ylabel('Numero de mortes')
plt.grid(True)

plt.savefig('numberdeathsBR.png')
```

# Brasil // SÃ£o Paulo // Nordeste

```python
bdBRcases = bdbrasilcases.iloc[:,2:bdbrasilcases.shape[1]]
bdNEcases = bdbrasilcases.loc[bdbrasilcases.iloc[:,0] == 'Nordeste',:].iloc[:,2:bdbrasilcases.shape[1]]
bdSPcases = bdbrasilcases.loc[bdbrasilcases.iloc[:,0] == 'SÃ£o Paulo',:].iloc[:,2:bdbrasilcases.shape[1]]

bdBRdeaths = bdbrasildeaths.iloc[:,2:bdbrasildeaths.shape[1]]
bdNEdeaths = bdbrasildeaths.loc[bdbrasildeaths.iloc[:,0] == 'Nordeste',:].iloc[:,2:bdbrasildeaths.shape[1]]
bdSPdeaths = bdbrasildeaths.loc[bdbrasildeaths.iloc[:,0] == 'SÃ£o Paulo',:].iloc[:,2:bdbrasildeaths.shape[1]]

cumulative_cases_br,cumulative_deaths_br = np.zeros(max(x)), np.zeros(max(x))
cumulative_cases_NE,cumulative_deaths_NE = np.zeros(max(x)), np.zeros(max(x))
cumulative_cases_SP,cumulative_deaths_SP = np.zeros(max(x)), np.zeros(max(x))

brcases,brdeaths,necases,nedeaths,spcases,spdeaths = 0,0,0,0,0,0
for i in range(0,(max(x))):
    cumulative_cases_br[i],cumulative_deaths_br[i] = bdBRcases.iloc[0,i] + brcases, bdBRdeaths.iloc[0,i] +brdeaths
    cumulative_cases_NE[i],cumulative_deaths_NE[i]  = bdNEcases.iloc[0,i] + necases, bdNEdeaths.iloc[0,i] + nedeaths
    cumulative_cases_SP[i],cumulative_deaths_SP[i] = bdSPcases.iloc[0,i] + spcases, bdSPdeaths.iloc[0,i] + spdeaths
    
    brcases,brdeaths = cumulative_cases_br[i],cumulative_deaths_br[i]
    necases,nedeaths = cumulative_cases_NE[i],cumulative_deaths_NE[i]
    spcases,spdeaths = cumulative_cases_SP[i],cumulative_deaths_SP[i] 
```
